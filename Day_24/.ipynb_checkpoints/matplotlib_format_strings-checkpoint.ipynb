<h1 align="center" style="color:green">Matplotlib Tutorial: pyplot.plot format strings and other options</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

### Simple plot

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'*')
```

### Line Style

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b--+')
```

### Marker Shapes

```python
# Same effect as 'b+' format string# but marker="D" makes it diamond - help(plt.plot)
markersize = 20
plt.plot(x,y,color='123456',marker="D",linestyle='--')

#     =============    ===============================
#     character        description
#     =============    ===============================
#     ``'.'``          point marker
#     ``','``          pixel marker
#     ``'o'``          circle marker
#     ``'v'``          triangle_down marker
#     ``'^'``          triangle_up marker
#     ``'<'``          triangle_left marker
#     ``'>'``          triangle_right marker
#     ``'1'``          tri_down marker
#     ``'2'``          tri_up marker
#     ``'3'``          tri_left marker
#     ``'4'``          tri_right marker
#     ``'s'``          square marker
#     ``'p'``          pentagon marker
#     ``'*'``          star marker
#     ``'h'``          hexagon1 marker
#     ``'H'``          hexagon2 marker
#     ``'+'``          plus marker
#     ``'x'``          x marker
#     ``'D'``          diamond marker
#     ``'d'``          thin_diamond marker
#     ``'|'``          vline marker
#     ``'_'``          hline marker
#     =============    ===============================
```

<h3 style="color:purple">markersize</h3>

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=40)

#     =============    ===============================
#     character        color
#     =============    ===============================
#     ``'b'``          blue
#     ``'g'``          green
#     ``'r'``          red
#     ``'c'``          cyan
#     ``'m'``          magenta
#     ``'y'``          yellow
#     ``'k'``          black
#     ``'w'``          white
#     =============    ===============================
```

<h3 style="color:purple">alpha property to control transparency of line chart</h3>

```python
plt.plot(x,y,'g<',alpha=1) # alpha can be specified on a scale 0 to 1
```
