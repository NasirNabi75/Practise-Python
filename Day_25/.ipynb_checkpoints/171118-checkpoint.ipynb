# seaborn plots

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset("tips")
tips.head()
```

## distplot

```python
x = np.random.normal(size=3)
print(x)
sns.distplot(x, kde=False)
```

```python
sns.distplot(tips["total_bill"],kde=False, bins= 100)
```

## jointplot

```python
# jointplot is more suitable for numerical data
sns.jointplot("total_bill", "tip", data= tips)

sns.jointplot("total_bill", "tip", data= tips, kind="scatter")
sns.jointplot("total_bill", "tip", data= tips, kind="reg")
sns.jointplot("total_bill", "tip", data= tips, kind="resid")
sns.jointplot("total_bill", "tip", data= tips, kind="kde")
sns.jointplot("total_bill", "tip", data= tips, kind="hex")
```

## pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex")
```

```python
sns.pairplot(tips, hue="sex", palette="RdBu_r" )
```

## rugplot

```python
sns.rugplot(tips["total_bill"])
```

## barplot

```python
sns.barplot(x="sex", y="total_bill", data=tips)
```

## countplot

```python
sns.countplot(x="sex", data=tips)
```

## boxplot

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
```

## stripplot

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="smoker", jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="smoker", jitter=False)
```
